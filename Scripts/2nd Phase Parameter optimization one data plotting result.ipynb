```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# FPSO-MLP

```python
feature=pd.read_csv('features_PSO_airfoil_w&c_one_data_first_feature_linear_train.csv')
print(feature)
actual_feature=pd.read_csv('actual_features_PSO_airfoil_w&c_one_data_first_feature_linear_train.csv')
print(actual_feature)
squared_error=pd.read_csv('airfoil_error_value_PSO_w&c_one_data_first_feature_linear_train.csv')
print(squared_error)
```

```python
fig=plt.figure(figsize=(20,7))
fig.suptitle('Parameter Gap and Square Error of Output (Airfoil)', fontsize=15)
ax1=fig.add_subplot(2,1,1)
ax1.plot(feature.iloc[:20,1], color='blue', label='Frequency')
ax1.plot(actual_feature.iloc[:20,1], color='red', label='Actual Frequency')
ax1.set_title('Actual and Estimated Parameter(X)')
ax1.legend()
#ax1.set_xlabel('Iteration', fontsize=15)
ax1.set_ylabel('value', fontsize=15)
ax2=fig.add_subplot(2,1,2)
ax2.plot(squared_error.iloc[:20,1], color='black', label='Squared Error')
ax2.legend()
#ax2.set_title('Squared Error')
ax2.set_xlabel('Iteration', fontsize=15)
ax2.set_ylabel('SE', fontsize=15)
plt.show()
fig.savefig('C:\\Users\\user\\Desktop\\Thesis\\FPSO_airfoil_featurevserror.jpg')
```
